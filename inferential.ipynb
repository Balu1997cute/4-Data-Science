```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(X):# we are creating one function

    # Calculating VIF
    vif = pd.DataFrame()# now we are passing input parameter for creating the table via automatically. This tabel is empty table 
    vif["variables"] = X.columns # now i want variable column name, for giving the dataset colmns in that variables
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])] 
# next i want to creae another values for respective varibale that's why i create VIF in LHS
    return(vif) # here i return the vif table
```

## Probability Density Function(PDF)

```python
# Density Polt

import seaborn as sns
x=dataset['ssc_p']
ax = sns.distplot(dataset['ssc_p'],kde=True)
```

```python
def get_pdf_probability(dataset,startrange,endrange):
    from matplotlib import pyplot
    from scipy.stats import norm
    import seaborn as sns
    ax = sns.distplot(dataset,kde=True,kde_kws={'color':'blue'},color='Green')
    pyplot.axvline(startrange,color='Red')
    pyplot.axvline(endrange,color='Red')
    # generate a sample
    sample = dataset
    # calculate parameters
    sample_mean =sample.mean()
    sample_std = sample.std()
    print('Mean=%.3f, Standard Deviation=%.3f' % (sample_mean, sample_std))
    # define the distribution
    dist = norm(sample_mean, sample_std)
    
    # sample probabilities for a range of outcomes
    values = [value for value in range(startrange, endrange)]
    probabilities = [dist.pdf(value) for value in values]    
    prob=sum(probabilities)
    print("The area between range({},{}):{}".format(startrange,endrange,sum(probabilities)))
    return prob
    
```

```python
get_pdf_probability(dataset['ssc_p'],60,60)
```

Diferrent Group(Male, Female) but same contion(salary)

```python
from scipy.stats import ttest_ind
dataset=dataset.dropna()
male = dataset[dataset['gender']=='M']['salary']
female = dataset[dataset['gender']=='F']['salary']
#print(male)
ttest_ind(male, female)
```

## One-Way Classification

```python
import scipy.stats as stats

stats.f_oneway(dataset['ssc_p'],dataset['hsc_p'],dataset['degree_p'])
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

#perform two-way ANOVA
model = ols('degree_p ~ C(ssc_p) + C(hsc_p) + C(ssc_p):C(hsc_p)', data=dataset).fit()
sm.stats.anova_lm(model, typ=2)
```
