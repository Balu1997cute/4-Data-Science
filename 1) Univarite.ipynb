```python
def freqtable(): # now we create a function  by adding  all the dataframe function
    frq=pd.DataFrame()
    frq["Unique_Values"]=dataset['ssc_p'].value_counts().index
    frq["Frequency"]=dataset['ssc_p'].value_counts().values
    frq["Relative_Frq"]=dataset['ssc_p'].value_counts().values/len(dataset['ssc_p'])*100
    frq["Cumulative"]=frq["Relative_Frq"].cumsum()
    return frq  
```

```python
def freqtable(columnName,dataset): # now i want to do automate so instead ssc_p i use any other column name. அதனால function alteration பண்ண்ரன்
    frq=pd.DataFrame()
    frq["Unique_Values"]=dataset[columnName].value_counts().index
    frq["Frequency"]=dataset[columnName].value_counts().values
    frq["Relative_Frq"]=dataset[columnName].value_counts().values/len(dataset[columnName])*100
    frq["Cumulative"]=frq["Relative_Frq"].cumsum()
    return frq 
```

```python
univariate=pd.DataFrame(index=["Mean","Median","Mode","Min","25%","50%","75%","Max","IQR","1.5rule","Lesser","Greater"],columns=quan)
for column in quan:  #here we are adding  the IQR, 1.5rule, Lesser and Greater in index and in table also
    print(column)
    univariate[column]["Mean"]=dataset[column].mean()
    univariate[column]["Median"]=dataset[column].median()
    univariate[column]["Mode"]=dataset[column].mode()[0]
    univariate[column]["Min"]=dataset[column].min()
    univariate[column]["25%"]=np.percentile(dataset[column],25)
    univariate[column]["50%"]=np.percentile(dataset[column],50)
    univariate[column]["75%"]=np.percentile(dataset[column],75)
    univariate[column]["Max"]=np.percentile(dataset[column],100)
    univariate[column]["IQR"]=univariate[column]["75%"]-univariate[column]["25%"]# (i.e) Q3-Q1
    univariate[column]["1.5rule"]=1.5*univariate[column]["IQR"]#1.5 * IQR
    univariate[column]["Lesser"]=univariate[column]["25%"]-univariate[column]["1.5rule"]  #Q1-1.5IQR
    univariate[column]["Greater"]=univariate[column]["75%"]+univariate[column]["1.5rule"] #Q3+1.5IQR
```

```python
def get_pdf_probability(dataset,startrange,endrange):
    from matplotlib import pyplot
    from scipy.stats import norm
    import seaborn as sns
    ax = sns.distplot(dataset,kde=True,kde_kws={'color':'blue'},color='Green')
    pyplot.axvline(startrange,color='Red')
    pyplot.axvline(endrange,color='Red')
    # generate a sample
    sample = dataset
    # calculate parameters
    sample_mean =sample.mean()
    sample_std = sample.std()
    print('Mean=%.3f, Standard Deviation=%.3f' % (sample_mean, sample_std))
    # define the distribution
    dist = norm(sample_mean, sample_std)
    
    # sample probabilities for a range of outcomes
    values = [value for value in range(startrange, endrange)]
    probabilities = [dist.pdf(value) for value in values]    
    prob=sum(probabilities)
    print("The area between range({},{}):{}".format(startrange,endrange,sum(probabilities)))
    return prob
    
```

```python
# Coverted to standard Normal Distribution
import seaborn as sns
mean=dataset['ssc_p'].mean()
std=dataset['ssc_p'].std()

values=[i for i in dataset['ssc_p']]

z_score=[((j-mean)/std) for j in values]

sns.distplot(z_score,kde=True)

sum(z_score)/len(z_score)
#z_score.std()
```

```python
def stdNBgraph(dataset):
        # Coverted to standard Normal Distribution
    import seaborn as sns
    mean=dataset.mean()
    std=dataset.std()

    values=[i for i in dataset]

    z_score=[((j-mean)/std) for j in values]

    sns.distplot(z_score,kde=True)

    sum(z_score)/len(z_score)
    #z_score.std()
```
