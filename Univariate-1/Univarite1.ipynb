```python
def freqtable(): # now we create a function  by adding  all the dataframe function
    frq=pd.DataFrame()
    frq["Unique_Values"]=dataset['ssc_p'].value_counts().index
    frq["Frequency"]=dataset['ssc_p'].value_counts().values
    frq["Relative_Frq"]=dataset['ssc_p'].value_counts().values/len(dataset['ssc_p'])*100
    frq["Cumulative"]=frq["Relative_Frq"].cumsum()
    return frq  
```

```python
def freqtable(columnName,dataset): # now i want to do automate so instead ssc_p i use any other column name. அதனால function alteration பண்ண்ரன்
    frq=pd.DataFrame()
    frq["Unique_Values"]=dataset[columnName].value_counts().index
    frq["Frequency"]=dataset[columnName].value_counts().values
    frq["Relative_Frq"]=dataset[columnName].value_counts().values/len(dataset[columnName])*100
    frq["Cumulative"]=frq["Relative_Frq"].cumsum()
    return frq 
```
